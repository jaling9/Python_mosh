### 03:55:34  Project 1: Automation with Python

````python
import openpyxl as xl # å¯¼å…¥åŒ…
from openpyxl.chart import BarChart,Reference # å¯¼å…¥å›¾è¡¨ç±»

wb = xl.load_workbook('transactions.xlsx') # å¯¼å…¥æ–‡ä»¶
sheet = wb['Sheet1'] # å–å‡ºSheetè¡¨
cell = sheet['a1'] # å•å…ƒæ ¼
cell = sheet.cell(1,1) # å•å…ƒæ ¼å¦å¤–ä¸€ç§æ–¹å¼
# print(cell.value) 

# print(sheet.max_row) # æŸ¥æ‰¾è¡Œæ•°
# print(sheet.max_column) # æŸ¥æ‰¾åˆ—æ•°

for row in range(2,sheet.max_row + 1): # + 1 æ˜¯å› ä¸ºpythonè®¡ç®—ç¬¬å‡ è¡Œçš„æ—¶å€™ä¸åŒ…å«â€œç¬¬å‡ â€è¡Œ
    # print(row)
    cell = sheet.cell(row,3) # å•å…ƒæ ¼ä½ç½®ç”¨åæ ‡å®šä½
    corrected_price = cell.value * 0.9
    corrected_price_cell = sheet.cell(row,4)
    corrected_price_cell.value = corrected_price

# æ¥ä¸‹æ¥å¼•ç”¨Referenceç±»åˆ›å»ºä¸€ä¸ªå€¼å¯¹è±¡
values = Reference(sheet, 
        min_row=2,
        max_row=sheet.max_row,
        min_col=4,
        max_col=4)

# æ¥ä¸‹æ¥å¼•ç”¨Barchartç±»åˆ›å»ºä¸€ä¸ªå›¾è¡¨å¯¹è±¡
chart = BarChart()
chart.add_data(values)
sheet.add_chart(chart, 'e2')

wb.save('transaction2.xlsx')

````
