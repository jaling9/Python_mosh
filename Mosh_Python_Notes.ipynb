### 02:26:21  Emoji Converter

```python
message = input(">")
words = message.split(' ')
emojis = {
    ":)":"ğŸ˜",
    ":(":"ğŸ˜’"
}
output = ""
for word in words:
    output = output + emojis.get(word,word) + " "
print(message)
print(output)
```
